```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data_eda/full_df.csv')
```

## Cairo Latitude: 30.033333, Longitude: 31.233334

```python
df_cairo = df.loc[(df['latitude'] > 29.8) & (df['latitude'] < 30.2) & (df['longitude'] > 31.1) & (df['longitude'] < 31.4)]
```

```python
df_cairo.shape
```

## graph

```python
# total population
df_cairo.plot(kind = 'scatter', 
        y = 'latitude', 
        x = 'longitude', 
        c = 'population_total',
        cmap = 'inferno', 
        figsize = (10,10), 
        s = 2);
```

```python
# man population
df_cairo.plot(kind = 'scatter', 
        y = 'latitude', 
        x = 'longitude', 
        c = 'population_man',
        cmap = 'inferno', 
        figsize = (10,10), 
        s = 2);
```

```python
# woman population
df_cairo.plot(kind = 'scatter', 
        y = 'latitude', 
        x = 'longitude', 
        c = 'population_woman',
        cmap = 'inferno', 
        figsize = (10,10), 
        s = 2);
```

```python
# woman rep population
df_cairo.plot(kind = 'scatter', 
        y = 'latitude', 
        x = 'longitude', 
        c = 'population_woman_rep',
        cmap = 'inferno', 
        figsize = (10,10), 
        s = 2);
```

```python
# children
df_cairo.plot(kind = 'scatter', 
        y = 'latitude', 
        x = 'longitude', 
        c = 'population_children',
        cmap = 'inferno', 
        figsize = (10,10), 
        s = 2);
```

```python
# youth
df_cairo.plot(kind = 'scatter', 
        y = 'latitude', 
        x = 'longitude', 
        c = 'population_youth',
        cmap = 'inferno', 
        figsize = (10,10), 
        s = 2);
```
