```python
# imports here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import time

pd.options.display.max_rows = 100


import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim , ArcGIS
from geopy.exc import GeocoderTimedOut
```

```python
url = 'https://www.citypopulation.de/en/egypt/greatercairo/'
res = requests.get(url)
```

```python
res.status_code
```

```python
soup = BeautifulSoup(res.content, 'lxml')
```

```python
len(soup.find('tbody'))
```

```python
t_body = soup.find('tbody')
```

```python
t_body
```

```python
# Attempt to scrape density data from the tbody
districts = []

for row in t_body.find_all('tr'):
    data = {}
    data['wiki_name'] = row.find('td').attrs['data-wiki']
    data['formal_name'] = row.find('span').text
    data['governorate'] = row.find('td', {'class': 'radm'}).text
    data['pop_density'] = row.find('td').attrs['data-density']
    data['2019_population'] = row.find('td', {'class': 'rpop prio1'}).text
    data['district_area_km2'] = row.find('td').attrs['data-area']
    
    districts.append(data)
    
df = pd.DataFrame(districts)
```
