```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/df_cairo.csv')
```

```python
df_house = pd.read_csv('../data/coord_no_zeros.csv')
```

## housing

```python
df_house.head(2)
```

```python
# Selecting only the city of Cairo real estate ads
df_house_s = df_house.loc[(df_house['Longitude'] > 29.8) & 
                  (df_house['Longitude'] < 30.2) & 
                  (df_house['Latitude'] > 31.1) & 
                  (df_house['Latitude'] < 31.4)]
```

```python
df_house_s.shape
```

```python
df_house_s.head(2)
```

## graph

```python
house_low = df_house_s.loc[df_house_s['base_price'] <= 2.500000e+04]
house_high = df_house_s.loc[df_house_s['base_price'] > 2.500000e+04]
```

```python
# latitude 
slum_lat = [30.0362,30.0666667,30.0833333,31.3,30.0711,30.0011,30.0645,30.1286111]
slum_lat_plot = [i for i in slum_lat if (i > 29.8) & (i < 30.2)]

# longitude
slum_lon = [31.2783,31.3,31.2166667,30.93,31.2300,31.2316,31.3816,31.2422222]
slum_lon_plot = [i for i in slum_lon if (i > 31.1) & (i < 31.4)]
```

```python
# total population
df.plot(kind = 'scatter', 
        y = 'latitude', 
        x = 'longitude', 
        c = 'population_total',
        cmap = 'binary', 
        figsize = (12,12), 
        s = 2)

# plt.figure(figsize = (8,8))

# housing price
plt.scatter(y = house_low['Longitude'], 
            x = house_low['Latitude'], 
            cmap = 'binary',
            marker = "*", 
            c = 'r',
            label = 'low housing price', 
            s =100)

plt.scatter(y = house_high['Longitude'], 
            x = house_high['Latitude'], 
            cmap = 'binary',
            marker = "+", 
            c = 'b',
            label = 'high housing price', 
            s =100)


# slum area
plt.scatter(y = slum_lat_plot, 
            x = slum_lon_plot, 
            cmap = 'gist_yarg', 
            marker = 'o',
            alpha = 0.5, 
            s = 1000, 
            c = 'brown', 
            label = 'slum')

plt.legend();
```
