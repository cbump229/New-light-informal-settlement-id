```python
locator= ArcGIS()
```

```python
df['lat'] = 0.0
df['lon'] = 0.0
```

```python
df.shape
```

```python
df.head()
```

```python
num=0
while num < len(df):
    try:
        location = locator.geocode(df["Location"][num].split(",")[2] + ",Cairo,Egypt")
        df.at[num, 'lat']= location.latitude
        df.at[num, 'lon']= location.longitude
    except AttributeError:
        pass
    except IndexError:
        pass
    except GeocoderTimedOut:
        pass
    if num %50 == 0:
        print(num)
    time.sleep(5)
    num+=1
    
```
