```python
density = pd.read_csv('CSVs/population_density_final.csv')
density.head()
```

```python
density['wiki_name'].value_counts()
```

#### Re-Mapping Density Names to Match DataBase

```python
density['db_name'] = density['wiki_name']
```

```python
density.head()
```

```python
df = density[['db_name', 'pop_density', '2019_population', 'district_area_km2']].copy()
```

```python
df.head()
```

```python
df['db_name'].replace('Bulaq','Bulaq Abo El Ela', inplace=True)
df['db_name'].replace('Al Nozha, Cairo', 'El Nozha', inplace=True)
df['db_name'].replace('Kasr El Nile', 'Garden City', inplace=True)
df['db_name'].replace('Maadi', 'Hay El Maadi', inplace=True)
df['db_name'].replace('Shubra', 'Hay Shobra', inplace=True)
df['db_name'].replace('Tora, Egypt', 'Hay Torah', inplace=True)

df['db_name'].replace('Al-Muqaṭṭam', 'Mokattam', inplace=True)
df['db_name'].replace('New Cairo', 'New Cairo City', inplace=True)
df['db_name'].replace('Heliopolis, Cairo', 'New Heliopolis', inplace=True)
df['db_name'].replace('El Shruq', 'Shorouk City', inplace=True)
```

```python
df['db_name'].value_counts()
```

```python
df.to_csv('CSVs/updated_density_list.csv', index=False)
```
